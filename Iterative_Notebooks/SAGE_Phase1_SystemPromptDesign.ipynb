```python

def run_prompt(technique_name, system_prompt, user_message, model="gpt-4o", temperature=0.3, max_tokens=2000):
    """
    Runs a prompt against the OpenAI API and returns a structured result.

    Args:
        technique_name: Name of the prompting technique (for logging)
        system_prompt: The system prompt to use
        user_message: The user message / query
        model: OpenAI model to use
        temperature: Controls randomness (lower = more deterministic)
        max_tokens: Maximum response length

    Returns:
        dict with technique, prompt, response, and usage metadata
    """
    print(f"\n{'='*70}")
    print(f"🔬 Technique: {technique_name}")
    print(f"{'='*70}")

    start_time = time.time()

    try:
        response = client.chat.completions.create(
            model=model,
            messages=[
                {"role": "system", "content": system_prompt},
                {"role": "user", "content": user_message}
            ],
            temperature=temperature,
            max_tokens=max_tokens
        )

        elapsed = time.time() - start_time
        result = {
            "technique": technique_name,
            "system_prompt": system_prompt,
            "user_message": user_message,
            "response": response.choices[0].message.content,
            "model": model,
            "temperature": temperature,
            "prompt_tokens": response.usage.prompt_tokens,
            "completion_tokens": response.usage.completion_tokens,
            "total_tokens": response.usage.total_tokens,
            "elapsed_seconds": round(elapsed, 2)
        }

        print(f"\n📊 Tokens: {result['prompt_tokens']} prompt + {result['completion_tokens']} completion = {result['total_tokens']} total")
        print(f"⏱️  Time: {result['elapsed_seconds']}s")
        print(f"\n📝 Response:\n{'-'*40}")
        print(result["response"])
        print(f"{'-'*40}")

        return result

    except Exception as e:
        print(f"❌ Error: {e}")
        return {"technique": technique_name, "error": str(e)}
```

```python
# =============================================================================
# TECHNIQUE 1: ZERO-SHOT PROMPTING
# =============================================================================
# No examples, no reasoning instructions. Just the role and the task.
# This is our baseline — we expect a shallow response that may miss
# policy dimensions or produce inconsistent formatting.
# =============================================================================

zero_shot_system = f"""You are a compliance policy assistant for TechNova Inc.
Answer questions based ONLY on the following policy documents.
If the policies do not contain enough information to answer, say so clearly.

POLICY DOCUMENTS:
{ALL_POLICIES}"""

zero_shot_user = TEST_QUERY

results["zero_shot"] = run_prompt(
    technique_name="Zero-Shot",
    system_prompt=zero_shot_system,
    user_message=zero_shot_user
)
```

```python

# =============================================================================
# TECHNIQUE 2: FEW-SHOT PROMPTING
# =============================================================================
# We provide 2 example Q&A pairs that demonstrate the desired output
# format (Answer, Citations, Risk Level, Reasoning). This teaches the
# model the structure, tone, and citation style we expect.
# =============================================================================

few_shot_system = f"""You are a compliance policy assistant for TechNova Inc.
Answer questions based ONLY on the following policy documents.
If the policies do not contain enough information to answer, say so clearly.

POLICY DOCUMENTS:
{ALL_POLICIES}

Here are examples of how you should respond:

---
EXAMPLE 1:
User: What is the data retention period for customer PII?
Assistant:
Answer:
Customer PII must be retained for no longer than 7 years after the end of the customer relationship, unless a longer period is required by law.

Citations:
- POL-DP-2025, Section 4.1

Risk Level: Low

Reasoning:
This is a straightforward policy lookup. Section 4.1 of the Data Privacy Policy directly states the 7-year retention period. No ambiguity or cross-policy analysis is required.

---
EXAMPLE 2:
User: Can an employee install a third-party project management tool on their company laptop?
Assistant:
Answer:
No. Employees must not install unauthorized software on company devices without prior approval from IT Security. The employee should submit a software approval request to the IT Security team before installation.

Citations:
- POL-IS-2025, Section 3.3

Risk Level: Medium

Reasoning:
Section 3.3 of the Information Security Policy explicitly prohibits unauthorized software installation. While the intent may be legitimate, unapproved software could introduce security vulnerabilities. The risk is Medium because the violation is procedural and can be remediated by obtaining proper approval.

---
Now answer the following question using the same format:"""

few_shot_user = TEST_QUERY

results["few_shot"] = run_prompt(
    technique_name="Few-Shot",
    system_prompt=few_shot_system,
    user_message=few_shot_user
)
```

```python

# =============================================================================
# TECHNIQUE 3: CHAIN-OF-THOUGHT (CoT) PROMPTING
# =============================================================================
# We instruct the model to reason step by step through the problem.
# This should surface more policy dimensions than zero-shot by forcing
# the model to systematically check each policy area.
# =============================================================================

cot_system = f"""You are a compliance policy assistant for TechNova Inc.
Answer questions based ONLY on the following policy documents.
If the policies do not contain enough information to answer, say so clearly.

POLICY DOCUMENTS:
{ALL_POLICIES}

When answering, think through the problem step by step:

Step 1: Identify which policy documents are relevant to this question.
Step 2: For each relevant policy, identify the specific sections that apply.
Step 3: Extract the key requirements, thresholds, and conditions from those sections.
Step 4: Determine whether the described scenario complies with or violates each requirement.
Step 5: Identify what approvals or actions are needed for compliance.
Step 6: Assess the overall compliance risk level (Low / Medium / High) based on the number and severity of issues found.
Step 7: Provide your final answer with citations to specific policy sections.

Show your reasoning for each step before giving the final answer."""

cot_user = TEST_QUERY

results["chain_of_thought"] = run_prompt(
    technique_name="Chain-of-Thought",
    system_prompt=cot_system,
    user_message=cot_user
)
```

```python

# =============================================================================
# TECHNIQUE 4: STEP-BACK PROMPTING
# =============================================================================
# Before answering the specific question, we ask the model to first
# consider the broader principles and risk categories that apply.
# This grounds the reasoning in abstractions before diving into details.
# =============================================================================

stepback_system = f"""You are a compliance policy assistant for TechNova Inc.
Answer questions based ONLY on the following policy documents.
If the policies do not contain enough information to answer, say so clearly.

POLICY DOCUMENTS:
{ALL_POLICIES}

Before answering the specific question, first step back and consider:

1. What are the general categories of compliance risk when an employee works internationally?
   Think about: employment law, tax implications, data sovereignty, information security, benefits coverage.

2. What are the general principles governing personal device usage in a corporate environment?
   Think about: device management, data protection, encryption, access controls.

3. What are the general principles governing cross-border handling of customer data?
   Think about: data transfer regulations, storage restrictions, adequacy frameworks.

After identifying these broader principles, apply them to the specific scenario to provide a grounded answer with citations."""

stepback_user = TEST_QUERY

results["step_back"] = run_prompt(
    technique_name="Step-Back",
    system_prompt=stepback_system,
    user_message=stepback_user
)
```

```python

# =============================================================================
# TECHNIQUE 5: ANALOGICAL PROMPTING
# =============================================================================
# We provide an analogous scenario that the model can use as a reasoning
# template, then ask it to apply similar logic to our test query.
# =============================================================================

analogical_system = f"""You are a compliance policy assistant for TechNova Inc.
Answer questions based ONLY on the following policy documents.
If the policies do not contain enough information to answer, say so clearly.

POLICY DOCUMENTS:
{ALL_POLICIES}

Consider this analogous compliance scenario and how it was resolved:

ANALOGOUS SCENARIO:
An employee wanted to work remotely from Canada for 20 days while attending a family event. They planned to use their company-issued laptop with VPN to access internal systems, but would not be handling any customer data.

Analysis of that scenario:
- Remote Work Policy (Section 4.1): 20 days is within the 30-day international limit → only manager approval needed.
- Information Security Policy (Section 4.1): Company laptop with VPN satisfies network security requirements.
- Data Privacy Policy: Not handling customer data, so cross-border data transfer rules do not apply.
- Risk Level: Low — single policy triggered, all requirements easily met.

Now apply similar reasoning to the following question. Note where the new scenario differs from the analogy and what additional policies or approvals those differences trigger."""

analogical_user = TEST_QUERY

results["analogical"] = run_prompt(
    technique_name="Analogical",
    system_prompt=analogical_system,
    user_message=analogical_user
)
```

```python

# =============================================================================
# TECHNIQUE 6: AUTO-CoT PROMPTING
# =============================================================================
# Instead of providing the reasoning steps ourselves (like in manual CoT),
# we ask the model to generate its own reasoning framework first,
# then follow that framework to answer the question.
# =============================================================================

autocot_system = f"""You are a compliance policy assistant for TechNova Inc.
Answer questions based ONLY on the following policy documents.
If the policies do not contain enough information to answer, say so clearly.

POLICY DOCUMENTS:
{ALL_POLICIES}

Before answering the question, do the following:

1. First, generate the reasoning steps you would need to follow to thoroughly analyze this compliance question. List these steps explicitly.

2. Then, follow your own reasoning steps one by one to arrive at your answer.

3. Finally, provide your conclusion with specific policy citations, required approvals, and a risk level assessment."""

autocot_user = TEST_QUERY

results["auto_cot"] = run_prompt(
    technique_name="Auto-CoT",
    system_prompt=autocot_system,
    user_message=autocot_user
)
```

```python

# =============================================================================
# TECHNIQUE 7: GENERATED KNOWLEDGE PROMPTING
# =============================================================================
# Two-step approach:
# Step 1 — Ask the model to generate relevant background knowledge
#           about the compliance domains involved.
# Step 2 — Feed that generated knowledge back as additional context
#           alongside the policies to answer the actual question.
# =============================================================================

# --- Step 1: Generate background knowledge ---
gk_step1_system = f"""You are a compliance policy assistant for TechNova Inc.
You have access to the following policy documents:

POLICY DOCUMENTS:
{ALL_POLICIES}

Before answering a user's question, generate relevant background knowledge
that would help analyze the compliance implications. Focus on:

1. Key compliance considerations for international remote work arrangements
2. Security implications of using personal devices for corporate data access
3. Data privacy considerations when handling customer data from foreign locations
4. How these different compliance domains typically interact and create compound risk

Generate this background knowledge based on the policy documents provided.
Do NOT answer the user's question yet — just produce the background knowledge."""

gk_step1_user = f"Generate background knowledge relevant to this scenario: {TEST_QUERY}"

print("\n" + "="*70)
print("🔬 Technique: Generated Knowledge — Step 1 (Knowledge Generation)")
print("="*70)

gk_step1_result = run_prompt(
    technique_name="Generated Knowledge - Step 1",
    system_prompt=gk_step1_system,
    user_message=gk_step1_user
)

# --- Step 2: Use generated knowledge to answer the question ---
generated_knowledge = gk_step1_result.get("response", "")

gk_step2_system = f"""You are a compliance policy assistant for TechNova Inc.
Answer questions based ONLY on the following policy documents.
If the policies do not contain enough information to answer, say so clearly.

POLICY DOCUMENTS:
{ALL_POLICIES}

ADDITIONAL BACKGROUND KNOWLEDGE (generated from policy analysis):
{generated_knowledge}

Using both the policy documents and the background knowledge above,
provide a comprehensive answer with specific citations, required approvals,
and a risk level assessment."""

gk_step2_user = TEST_QUERY

results["generated_knowledge"] = run_prompt(
    technique_name="Generated Knowledge - Step 2 (Final Answer)",
    system_prompt=gk_step2_system,
    user_message=gk_step2_user
)

# Also store step 1 for documentation
results["generated_knowledge_step1"] = gk_step1_result
```

```python
# =============================================================================
# PART 2: ADVANCED SYSTEM PROMPT
# =============================================================================
# Add these cells after your existing foundational techniques cells.
# =============================================================================

# =============================================================================
# INITIAL SYSTEM PROMPT (First Draft)
# =============================================================================
# Combined from the best elements of each foundational technique:
# - Zero-Shot: base role + grounding constraint
# - Few-Shot: structured output format (Answer/Citations/Risk/Reasoning)
# - CoT: 7-step reasoning framework
# - Step-Back: broad risk category awareness
# - Analogical: comparative reasoning approach
# - Generated Knowledge: deep domain context
# =============================================================================

INITIAL_SYSTEM_PROMPT = f"""You are a compliance policy assistant for TechNova Inc.
Your role is to answer policy and compliance questions accurately, with full
citations, based ONLY on the provided policy documents.

STRICT RULES:
1. Use ONLY the provided policy documents to answer. Do not use outside knowledge.
2. Do not invent or fabricate policy sections, citations, or requirements.
3. If the policies do not contain sufficient information, say so explicitly.
4. Do not speculate beyond what the policy text states.
5. When a scenario involves multiple policies, analyze ALL relevant policies.

POLICY DOCUMENTS:
{ALL_POLICIES}

REASONING PROCESS:
When answering a question, follow these steps:
Step 1: Identify all policy documents relevant to the question.
Step 2: For each relevant policy, locate the specific sections that apply.
Step 3: Extract key requirements, thresholds, and conditions.
Step 4: Determine whether the scenario complies with or violates each requirement.
Step 5: Identify all approvals or actions needed for compliance.
Step 6: Assess the overall risk level (Low / Medium / High) based on the number
        and severity of compliance gaps.

RESPONSE FORMAT:
Always structure your response exactly as follows:

Answer:
[Comprehensive answer addressing all aspects of the question]

Citations:
- [Document ID], [Section number] — [Brief description of what it states]

Risk Level: [Low / Medium / High]

Reasoning:
[2-4 sentence summary of how you arrived at your answer and risk assessment]
"""

print("✅ Initial System Prompt defined")
print(f"   Length: {len(INITIAL_SYSTEM_PROMPT.split())} words")

results["initial_system_prompt"] = run_prompt(
    technique_name="Initial System Prompt (First Draft)",
    system_prompt=INITIAL_SYSTEM_PROMPT,
    user_message=TEST_QUERY
)
```

```python

# =============================================================================
# ADVANCED TECHNIQUE 1: DECOMPOSITION
# =============================================================================
# Instead of one monolithic prompt, we decompose the task into 3 stages:
#   Stage 1: Intent Classification
#   Stage 2: Policy Analysis (extract relevant sections + compliance check)
#   Stage 3: Risk Assessment & Final Response
# This mirrors the pipeline architecture from our PRD.
# =============================================================================

print("\n" + "="*70)
print("🔬 Advanced Technique 1: DECOMPOSITION (3-Stage Pipeline)")
print("="*70)

# --- Stage 1: Intent Classification ---
decomp_stage1_system = """You are an intent classifier for a compliance policy assistant.

Given a user query, classify it into exactly ONE of these categories:
- policy_question: The user is asking about a specific policy or procedure.
- risk_assessment: The user is describing a scenario and wants compliance risk evaluated.
- out_of_scope: The query is not related to company policies.

Respond with ONLY the category name, nothing else."""

decomp_stage1_result = run_prompt(
    technique_name="Decomposition — Stage 1: Intent Classification",
    system_prompt=decomp_stage1_system,
    user_message=TEST_QUERY
)

detected_intent = decomp_stage1_result.get("response", "").strip().lower()
print(f"\n🏷️  Detected Intent: {detected_intent}")

# --- Stage 2: Policy Analysis ---
decomp_stage2_system = f"""You are a policy analysis engine for TechNova Inc.
Your job is to identify ALL relevant policy sections for a given scenario
and extract the specific requirements that apply.

POLICY DOCUMENTS:
{ALL_POLICIES}

For the user's scenario, provide:
1. A list of every relevant policy section (Document ID + Section number).
2. For each section, state the specific requirement that applies.
3. For each requirement, state whether the scenario COMPLIES, VIOLATES, or REQUIRES ACTION.

Do NOT provide a final answer or risk assessment yet. Only perform the policy analysis."""

decomp_stage2_result = run_prompt(
    technique_name="Decomposition — Stage 2: Policy Analysis",
    system_prompt=decomp_stage2_system,
    user_message=TEST_QUERY
)

policy_analysis = decomp_stage2_result.get("response", "")

# --- Stage 3: Risk Assessment & Final Response ---
decomp_stage3_system = f"""You are a compliance risk assessor for TechNova Inc.

Based on the policy analysis below, synthesize a final response for the user.

POLICY ANALYSIS FROM PREVIOUS STAGE:
{policy_analysis}

Provide your response in this exact format:

Answer:
[Comprehensive answer addressing all aspects of the question]

Citations:
- [Document ID], [Section number] — [Brief description]

Risk Level: [Low / Medium / High]

Reasoning:
[2-4 sentence summary explaining the risk assessment based on the policy analysis]"""

decomp_stage3_result = run_prompt(
    technique_name="Decomposition — Stage 3: Risk Assessment & Response",
    system_prompt=decomp_stage3_system,
    user_message=TEST_QUERY
)

results["decomposition"] = {
    "technique": "Decomposition (3-Stage Pipeline)",
    "stage1_intent": decomp_stage1_result,
    "stage2_analysis": decomp_stage2_result,
    "stage3_response": decomp_stage3_result,
    "final_response": decomp_stage3_result.get("response", ""),
    "total_tokens": sum([
        decomp_stage1_result.get("total_tokens", 0),
        decomp_stage2_result.get("total_tokens", 0),
        decomp_stage3_result.get("total_tokens", 0)
    ]),
    "total_elapsed": round(sum([
        decomp_stage1_result.get("elapsed_seconds", 0),
        decomp_stage2_result.get("elapsed_seconds", 0),
        decomp_stage3_result.get("elapsed_seconds", 0)
    ]), 2)
}

print(f"\n📊 Decomposition Total: {results['decomposition']['total_tokens']} tokens, {results['decomposition']['total_elapsed']}s")
```

```python

# =============================================================================
# ADVANCED TECHNIQUE 2: ENSEMBLING
# =============================================================================
# Run 3 different prompt variations (different emphasis/framing) against
# the same query. Then aggregate the best elements into a final answer.
# This reduces the chance of any single prompt's blind spots affecting output.
# =============================================================================

print("\n" + "="*70)
print("🔬 Advanced Technique 2: ENSEMBLING (3 Prompt Variants)")
print("="*70)

# Variant A: Emphasizes legal/regulatory perspective
ensemble_a_system = f"""You are a compliance policy assistant for TechNova Inc.
Focus on LEGAL AND REGULATORY implications when analyzing the scenario.
Use ONLY the provided policy documents.

POLICY DOCUMENTS:
{ALL_POLICIES}

Provide: Answer, Citations, Risk Level, and Reasoning."""

# Variant B: Emphasizes information security perspective
ensemble_b_system = f"""You are a compliance policy assistant for TechNova Inc.
Focus on INFORMATION SECURITY AND DATA PROTECTION implications when analyzing the scenario.
Use ONLY the provided policy documents.

POLICY DOCUMENTS:
{ALL_POLICIES}

Provide: Answer, Citations, Risk Level, and Reasoning."""

# Variant C: Emphasizes operational/HR perspective
ensemble_c_system = f"""You are a compliance policy assistant for TechNova Inc.
Focus on OPERATIONAL AND HR implications when analyzing the scenario.
Use ONLY the provided policy documents.

POLICY DOCUMENTS:
{ALL_POLICIES}

Provide: Answer, Citations, Risk Level, and Reasoning."""

ensemble_a = run_prompt("Ensembling — Variant A (Legal Focus)", ensemble_a_system, TEST_QUERY)
ensemble_b = run_prompt("Ensembling — Variant B (Security Focus)", ensemble_b_system, TEST_QUERY)
ensemble_c = run_prompt("Ensembling — Variant C (HR/Ops Focus)", ensemble_c_system, TEST_QUERY)

# Aggregation step: feed all 3 responses to a final synthesizer
ensemble_agg_system = f"""You are a compliance policy assistant for TechNova Inc.

Three different analysts have reviewed the same compliance scenario from different
perspectives. Their analyses are provided below.

ANALYST A (Legal & Regulatory Focus):
{ensemble_a.get('response', '')}

ANALYST B (Information Security & Data Protection Focus):
{ensemble_b.get('response', '')}

ANALYST C (Operational & HR Focus):
{ensemble_c.get('response', '')}

Synthesize these three analyses into a single, comprehensive response that captures
the key findings from all perspectives. Resolve any contradictions by favoring
the most conservative (risk-averse) interpretation.

Use this format:
Answer:
[Comprehensive synthesized answer]

Citations:
- [All unique citations from the three analyses]

Risk Level: [The highest risk level among the three]

Reasoning:
[Summary that integrates insights from all three perspectives]"""

ensemble_final = run_prompt(
    "Ensembling — Final Aggregation",
    ensemble_agg_system,
    TEST_QUERY
)

results["ensembling"] = {
    "technique": "Ensembling (3 Variants + Aggregation)",
    "variant_a": ensemble_a,
    "variant_b": ensemble_b,
    "variant_c": ensemble_c,
    "aggregated": ensemble_final,
    "final_response": ensemble_final.get("response", ""),
    "total_tokens": sum([
        ensemble_a.get("total_tokens", 0),
        ensemble_b.get("total_tokens", 0),
        ensemble_c.get("total_tokens", 0),
        ensemble_final.get("total_tokens", 0)
    ])
}
```

```python

# =============================================================================
# ADVANCED TECHNIQUE 3: SELF-CONSISTENCY
# =============================================================================
# Run the SAME Initial System Prompt 3 times and check if the answers
# agree on key dimensions: risk level, required approvals, cited sections.
# Disagreements indicate prompt instability that needs fixing.
# =============================================================================

print("\n" + "="*70)
print("🔬 Advanced Technique 3: SELF-CONSISTENCY (3 Identical Runs)")
print("="*70)

sc_runs = []
for i in range(3):
    run = run_prompt(
        technique_name=f"Self-Consistency — Run {i+1}",
        system_prompt=INITIAL_SYSTEM_PROMPT,
        user_message=TEST_QUERY,
        temperature=0.5  # Slightly higher temp to allow natural variation
    )
    sc_runs.append(run)

# Compare the 3 runs
print("\n" + "="*70)
print("📊 SELF-CONSISTENCY COMPARISON")
print("="*70)

for i, run in enumerate(sc_runs):
    resp = run.get("response", "")
    # Extract risk level from response
    risk = "Unknown"
    for line in resp.split("\n"):
        if "risk level" in line.lower():
            risk = line.strip()
            break
    print(f"\nRun {i+1}: {risk}")
    print(f"  Response length: {len(resp.split())} words")

results["self_consistency"] = {
    "technique": "Self-Consistency (3 Runs)",
    "runs": sc_runs,
    "total_tokens": sum([r.get("total_tokens", 0) for r in sc_runs])
}
```

```python

# =============================================================================
# ADVANCED TECHNIQUE 4: UNIVERSAL SELF-CONSISTENCY
# =============================================================================
# After generating 3 responses, ask the model ITSELF to evaluate which
# response is most consistent, complete, and accurate — and produce a
# final unified answer.
# =============================================================================

print("\n" + "="*70)
print("🔬 Advanced Technique 4: UNIVERSAL SELF-CONSISTENCY")
print("="*70)

usc_system = f"""You are a compliance quality reviewer for TechNova Inc.

Three responses were generated for the same compliance question. Your job is to:

1. Compare all three responses for consistency, completeness, and accuracy.
2. Identify any contradictions or missing information across the responses.
3. Select the most accurate and complete elements from each response.
4. Produce a single, definitive answer that represents the best synthesis.

RESPONSE 1:
{sc_runs[0].get('response', '')}

RESPONSE 2:
{sc_runs[1].get('response', '')}

RESPONSE 3:
{sc_runs[2].get('response', '')}

First, briefly note any inconsistencies you found across the three responses.
Then provide the final unified answer in this format:

Consistency Analysis:
[Brief note on what was consistent and what differed]

Answer:
[Definitive synthesized answer]

Citations:
- [All verified citations]

Risk Level: [Justified risk level]

Reasoning:
[Summary integrating the strongest reasoning from all three responses]"""

usc_result = run_prompt(
    technique_name="Universal Self-Consistency — Evaluation & Synthesis",
    system_prompt=usc_system,
    user_message=TEST_QUERY
)

results["universal_self_consistency"] = {
    "technique": "Universal Self-Consistency",
    "evaluation": usc_result,
    "final_response": usc_result.get("response", ""),
    "total_tokens": results["self_consistency"]["total_tokens"] + usc_result.get("total_tokens", 0)
}
```

```python

# =============================================================================
# ADVANCED TECHNIQUE 5: SELF-CRITICISM
# =============================================================================
# Two-step process:
# Step 1: Generate a response using the Initial System Prompt
# Step 2: Ask the model to critically review its own response for
#          unsupported claims, missing citations, and reasoning gaps
# Step 3: Produce a corrected final response
# =============================================================================

print("\n" + "="*70)
print("🔬 Advanced Technique 5: SELF-CRITICISM (Generate → Critique → Revise)")
print("="*70)

# Step 1: Generate initial response (reuse the initial system prompt result)
initial_response = results["initial_system_prompt"].get("response", "")

print(f"\n--- Using Initial System Prompt response as the base for criticism ---")

# Step 2: Critique
critic_system = f"""You are a compliance quality auditor for TechNova Inc.
Your job is to critically review a response generated by a compliance assistant.

POLICY DOCUMENTS (ground truth):
{ALL_POLICIES}

ORIGINAL QUESTION:
{TEST_QUERY}

RESPONSE TO REVIEW:
{initial_response}

Evaluate the response against these criteria:

1. GROUNDEDNESS: Is every claim in the response directly supported by the policy
   documents? Identify any claims that are NOT supported by the text.

2. CITATION ACCURACY: Are all cited policy sections correct? Does the cited section
   actually say what the response claims it says?

3. COMPLETENESS: Are there any relevant policy sections that the response missed?
   Check all three policy documents systematically.

4. RISK ASSESSMENT: Is the risk level justified by the evidence? Is it too high
   or too low given the number and severity of compliance issues?

5. AMBIGUITY HANDLING: Did the response appropriately flag any ambiguous or
   unclear policy language instead of presenting assumptions as facts?

Provide your critique in this format:

Groundedness Issues:
[List any unsupported claims, or state "None found"]

Citation Errors:
[List any incorrect citations, or state "None found"]

Missing Coverage:
[List any policy sections that should have been referenced but were not]

Risk Level Assessment:
[Is the risk level appropriate? Should it be higher or lower? Why?]

Suggested Improvements:
[Specific changes that would improve the response]"""

critic_result = run_prompt(
    technique_name="Self-Criticism — Step 2: Critique",
    system_prompt=critic_system,
    user_message="Review the response above and provide your critique."
)

critique = critic_result.get("response", "")

# Step 3: Revise based on critique
revise_system = f"""You are a compliance policy assistant for TechNova Inc.

You previously generated a response to a compliance question, and a quality auditor
has reviewed it and identified issues. Your job is to produce a REVISED response
that addresses all the critique points.

POLICY DOCUMENTS:
{ALL_POLICIES}

YOUR ORIGINAL RESPONSE:
{initial_response}

AUDITOR'S CRITIQUE:
{critique}

Produce a revised response that:
- Fixes all groundedness issues (remove unsupported claims)
- Corrects all citation errors
- Adds any missing policy coverage
- Adjusts the risk level if the critique suggests it
- Flags any genuine ambiguities in the policies

Use this format:

Answer:
[Revised comprehensive answer]

Citations:
- [Corrected and complete citations]

Risk Level: [Adjusted if needed]

Reasoning:
[Updated reasoning addressing the critique points]"""

revise_result = run_prompt(
    technique_name="Self-Criticism — Step 3: Revised Response",
    system_prompt=revise_system,
    user_message=TEST_QUERY
)

results["self_criticism"] = {
    "technique": "Self-Criticism (Generate → Critique → Revise)",
    "original_response": initial_response,
    "critique": critic_result,
    "revised_response": revise_result,
    "final_response": revise_result.get("response", ""),
    "total_tokens": (
        results["initial_system_prompt"].get("total_tokens", 0) +
        critic_result.get("total_tokens", 0) +
        revise_result.get("total_tokens", 0)
    )
}
```

```python

# =============================================================================
# FINAL COMPARISON: ALL TECHNIQUES
# =============================================================================

print("\n" + "="*70)
print("📊 COMPLETE TECHNIQUE COMPARISON — FOUNDATIONAL + ADVANCED")
print("="*70)

all_techniques = [
    ("zero_shot", "Zero-Shot"),
    ("few_shot", "Few-Shot"),
    ("chain_of_thought", "Chain-of-Thought"),
    ("step_back", "Step-Back"),
    ("analogical", "Analogical"),
    ("auto_cot", "Auto-CoT"),
    ("generated_knowledge", "Generated Knowledge"),
    ("initial_system_prompt", "Initial System Prompt"),
    ("decomposition", "Decomposition"),
    ("ensembling", "Ensembling"),
    ("self_consistency", "Self-Consistency"),
    ("universal_self_consistency", "Universal Self-Consistency"),
    ("self_criticism", "Self-Criticism"),
]

print(f"\n{'Technique':<35} {'Total Tokens':<15} {'Category':<15}")
print("-" * 65)

for key, name in all_techniques:
    if key in results:
        r = results[key]
        tokens = r.get("total_tokens", 0)
        category = "Foundational" if key in ["zero_shot","few_shot","chain_of_thought","step_back","analogical","auto_cot","generated_knowledge"] else "Advanced"
        if key == "initial_system_prompt":
            category = "Initial Draft"
        print(f"{name:<35} {tokens:<15} {category:<15}")
```
